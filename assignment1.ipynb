```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
new_image = cv2.imread('DATA/test1.jpg')
```

```python
plt.imshow(new_image)
```

```python
#converting to rgb
new_img = cv2.cvtColor(new_image,cv2.COLOR_BGR2RGB)
plt.imshow(new_img)
```

```python
#flip image upside down
flip_img = cv2.flip(new_img,-1)
plt.imshow(flip_img)
```

```python
#drawing rectangle
cv2.rectangle(new_img,pt1 = (600,600),pt2 = (400,400),color = (0,255,0),thickness=8)
plt.imshow(new_img)
```

```python
#drawing triangle
new_img = cv2.cvtColor(new_image,cv2.COLOR_BGR2RGB)
p1 = (400,200)
p2 = (600,200)
p3 =(600,400)
cv2.line(new_img, p1 , p2 , (255, 0, 0), 3)
cv2.line(new_img, p2, p3, (255, 0, 0), 3)
cv2.line(new_img, p1, p3, (255, 0, 0), 3)
plt.imshow(new_img)
```

```python
def draw_circle(event,x,y,flag,param):
    
    if event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(new_img,(x,y),100,(255,0,0),3)
        
cv2.namedWindow(winname = "my_image")

cv2.setMouseCallback("my_image",draw_circle)
 
while True:
    cv2.imshow("my_image",new_img)
    if cv2.waitKey(1) & 0xFF == 27:
        break
cv2.destroyAllWindows()
```
