```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread("DATA/CR7.jpg")
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.imread("DATA/goat.jpg")
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
new_img1 = cv2.resize(img1,(800,800))
new_img2 = cv2.resize(img2,(800,800))
plt.imshow(new_img1)
```

```python
plt.imshow(new_img2)
```

```python
blended = cv2.addWeighted(src1= new_img1,alpha=0.9,src2=new_img2,beta=0.3,gamma=0)
plt.imshow(blended)
```

```python
#overlay of images (no blending)
img1 = cv2.imread("DATA/CR7.jpg")
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.imread("DATA/goat.jpg")
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
img1n = cv2.resize(img1,(800,800))
plt.imshow(img1n)
```

```python
img2n = cv2.resize(img2,(400,400))
plt.imshow(img2n)
```

```python
larger_img = img1n
smaller_img = img2n
```

```python
x_spot = 0
y_spot = 0
```

```python
x_end = x_spot + smaller_img.shape[1]
y_end = y_spot + smaller_img.shape[0]
```

```python
larger_img[y_spot:y_end,x_spot:x_end] = smaller_img
plt.imshow(larger_img)
```
